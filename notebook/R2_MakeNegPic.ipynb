###读取指定目录下的负样本文件（原图），并将文件名输出到txt文档中

```python
import re

pattern = re.compile(r'.*list.*|.*txt')

def GetFileList(direct, fileList):
    newdir = direct
    if os.path.isfile(direct):
#         fileList.append(direct.decode('gbk'))
        fileList.append(direct)
    elif os.path.isdir(direct):
        for s in os.listdir(direct):
            #如果需要忽略某些文件夹，使用以下代码
            if s == ".DS_Store" or s == "Thumbs.db" or s == 'neg_sub_url.txt':
                continue
            if pattern.match(s):
                print s
                continue
            newdir = os.path.join(direct,s)
            GetFileList(newdir, fileList)
        return fileList
```

```python
negdir = '/Users/chickenchen/Data/negpic/'

txt = GetFileList(negdir, [])
output = open(negdir+'neg_raw_url.txt', 'w')
for t in txt:
    output.writelines(t+'\n')
output.close()
```

###目录操作函数

```python
import os

def mkdir(path):
#     # 去除首位空格
#     path=path.strip()
    # 去除尾部 / 符号
    path=path.rstrip("/")
 
    # 判断路径是否存在
    # 存在     True
    # 不存在   False
    isExists=os.path.exists(path)
 
    # 判断结果
    if not isExists:
        # 如果不存在则创建目录
        print path+' 创建成功'
        # 创建目录操作函数
        os.makedirs(path)
        return True
    else:
        # 如果目录存在则不创建，并提示目录已存在
        print path+' 目录已存在'
        return False
 
```
